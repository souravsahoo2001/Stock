# StockVisualisation(APPLE Vs MICROSOFT Vs AMAZON)

```python
#importing modules
import numpy as np
import datetime
import pandas_datareader as pdr
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
start = datetime.datetime(2021,1,1)
end = datetime.datetime(2022,1,1)
```

```python
apple = pdr.DataReader('AAPL','yahoo',start,end)
amazon = pdr.DataReader('AMZN','yahoo',start,end)
microsoft = pdr.DataReader('MSFT','yahoo',start,end)
```

### Plot between Open Prices of three stocks

```python
microsoft['Open'].plot(label= 'Microsoft Open Price')
amazon['Open'].plot(label= 'Amazon Open Price')
apple['Open'].plot(label= "Apple Open Price",figsize= (15,7))
plt.legend()
plt.ylabel('Stock Price In USD',fontsize= 15)
plt.xlabel('Date',fontsize= 15)
plt.title('Open Prices Of Stocks in 2021-2022', fontsize= 18)
plt.grid(color = 'black',linestyle= ':', linewidth= 0.5)
plt.show()
```

### Plot Between Close Prices of three stocks

```python
microsoft['Close'].plot(label= 'Microsoft Close Price')
amazon['Close'].plot(label= 'Amazon Close Price')
apple['Close'].plot(label= "Apple Close Price",figsize= (15,7))
plt.legend()
plt.ylabel('Stock Price In USD',fontsize= 15)
plt.xlabel('Date',fontsize= 15)
plt.title('Close Prices Of Stocks in 2021-2022', fontsize= 18)
plt.grid(color = 'black',linestyle= ':', linewidth= 0.5)
plt.show()
```

##### We can analyse from the above plot clearly that microsoft completely out run the other two companies through out the year.But if we observe the graph closely the difference between the open price and close price value for a particular company is very less for the same day.

#####                                                       Amazon has dominated over apple through out the year, but the at the end of december the stock price of apple gradually increased, and by the end of december it crossed over Amazon. 


## Comparision of volume of the stocks through out the year

```python
microsoft['Volume'].plot(label= 'Microsoft')
amazon['Volume'].plot(label= 'Amazon')
apple['Volume'].plot(label= "Apple",figsize= (15,7))
plt.legend()
plt.ylabel('Volume',fontsize= 15)
plt.xlabel('Date',fontsize= 15)
plt.title('Volume of Traded', fontsize= 18)
plt.show()
```

```python
#To find the particular date of maximum volume we use iloc and argmax function
amazon.iloc[[amazon['Volume'].argmax()]]
```

##### So from the above chart we can notice a huge spike of amazon around august(July 30 to be accurate) of 2021.That means at that particular of time the traded stock of amazon was maximum.

## Moving Average Of Stock

###### A moving average (MA) is a stock indicator commonly used in technical analysis, used to help smooth out price data by creating a constantly updated average price. A rising moving average indicates that the security is in an uptrend, while a declining moving average indicates a downtrend.

```python
#Use of rolling function to find the moving average of a stock
microsoft['Open'].plot(label ='No Moving Average', figsize= (15,7))
microsoft['MA50'] = microsoft['Open'].rolling(50).mean()
microsoft['MA50'].plot(label= 'MA50')
plt.grid(color = 'black',linestyle= ':', linewidth= 0.5)
plt.legend(loc= 'lower right')
plt.show()
```
