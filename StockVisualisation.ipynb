# StockVisualisation(APPLE Vs MICROSOFT Vs AMAZON)

```python
#importing modules
import numpy as np
import datetime
import pandas_datareader as pdr
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
start = datetime.datetime(2021,1,1)
end = datetime.datetime(2022,1,1)
```

```python
apple = pdr.DataReader('AAPL','yahoo',start,end)
amazon = pdr.DataReader('AMZN','yahoo',start,end)
microsoft = pdr.DataReader('MSFT','yahoo',start,end)
```

### Plot between Open Prices of three stocks

```python
microsoft['Open'].plot(label= 'Microsoft Open Price')
amazon['Open'].plot(label= 'Amazon Open Price')
apple['Open'].plot(label= "Apple Open Price",figsize= (15,7))
plt.legend()
plt.ylabel('Stock Price In USD',fontsize= 15)
plt.xlabel('Date',fontsize= 15)
plt.title('Open Prices Of Stocks in 2021-2022', fontsize= 18)
plt.grid(color = 'black',linestyle= ':', linewidth= 0.5)
plt.show()
```

### Plot Between Close Prices of three stocks

```python
microsoft['Close'].plot(label= 'Microsoft Close Price')
amazon['Close'].plot(label= 'Amazon Close Price')
apple['Close'].plot(label= "Apple Close Price",figsize= (15,7))
plt.legend()
plt.ylabel('Stock Price In USD',fontsize= 15)
plt.xlabel('Date',fontsize= 15)
plt.title('Close Prices Of Stocks in 2021-2022', fontsize= 18)
plt.grid(color = 'black',linestyle= ':', linewidth= 0.5)
plt.show()
```

##### We can analyse from the above plot clearly that microsoft completely out run the other two companies through out the year.But if we observe the graph closely the difference between the open price and close price value for a particular company is very less for the same day.

#####                                                       Amazon has dominated over apple through out the year, but the at the end of december the stock price of apple gradually increased, and by the end of december it crossed over Amazon.

## Moving Average Of Stock

###### A moving average (MA) is a stock indicator commonly used in technical analysis, used to help smooth out price data by creating a constantly updated average price. A rising moving average indicates that the security is in an uptrend, while a declining moving average indicates a downtrend.

```python
#Use of rolling function to find the moving average of a stock
microsoft['Open'].plot(label ='No Moving Average', figsize= (15,7))
microsoft['MA50'] = microsoft['Open'].rolling(50).mean()
microsoft['MA50'].plot(label= 'MA50')
plt.grid(color = 'black',linestyle= ':', linewidth= 0.5)
plt.legend(loc= 'lower right')
plt.show()
```

# Daily Return (Daily Percentage Change)


#### Basically this just informs  your percent gain (or loss) if you bought the stock on day and then sold it the next day. Daily percentage change is defined by the following formula:

#### Daily return = (Today prices - Yesterday price) / Yesterday price
                    
                
  
 
 


$ Daily Return = \frac{Today Price}{Yesterday Price}-1$ $ (r_t = \frac{p_t}{p_t-1}-1) $

##### This defines r_t (return at time t) as equal to the price at time t divided by the price at time t-1 (the previous day) minus 1. While this isn't necessarily helpful for attempting to predict future values of the stock, its very helpful in analyzing the volatility of the stock. If daily returns have a wide distribution, the stock is more volatile from one day to the next.

```python
#Adding a new column "Daily_return" in the data 
microsoft['Daily_return']=microsoft['Close']/microsoft['Close'].shift(1)-1
amazon['Daily_return']=amazon['Close']/amazon['Close'].shift(1)-1
apple['Daily_return']=apple['Close']/apple['Close'].shift(1)-1


```

```python
microsoft.head()
```

```python
microsoft['Daily_return'].plot(figsize= (15,7))
plt.legend()
plt.show()
```

```python
amazon.head()
```

```python
amazon['Daily_return'].plot(figsize= (15,7))
plt.legend()
plt.show()
```

```python
apple.head()
```

```python
apple['Daily_return'].plot(figsize= (15,7))
plt.legend()
plt.show()
```

### Comparing all three stock's daily return at one place using subplot function

```python
plt.figure(figsize= (15,10))
plt.subplot(3,1,1)
plt.plot(microsoft['Daily_return'],label = 'Microsoft')
plt.ylabel('Daily Return',fontsize= 13)
plt.xlabel('Date',fontsize= 13)
plt.title('Microsoft Daily Return',fontsize= 15)
plt.legend()
plt.show()


plt.figure(figsize= (15,10))
plt.subplot(3,1,2)
plt.plot(amazon['Daily_return'],label = 'Amazon',color= 'orange')
plt.ylabel('Daily Return',fontsize= 13)
plt.xlabel('Date',fontsize= 13)
plt.title('Amazon Daily Return',fontsize= 15)
plt.legend()
plt.show()


plt.figure(figsize= (15,10))
plt.subplot(3,1,3)
plt.plot(apple['Daily_return'],label = 'Apple',color= 'black')
plt.ylabel('Daily Return',fontsize= 13)
plt.xlabel('Date',fontsize= 13)
plt.title('Apple Daily Return',fontsize= 15)
plt.legend()
plt.show()
```

### Comparing the Volatility(Daily Percentage Change) through Piechart

```python
md = (microsoft['Daily_return'].sum())
amd = (amazon['Daily_return'].sum())
apd = (apple['Daily_return'].sum())
print(md,amd,apd)
```

```python
from matplotlib import style
plt.figure(figsize= (5,5))
stocks = ['Microsoft','Amazon','Apple']
values = (md,amd,apd)
plt.pie(values,labels= stocks,autopct= '%.1f%%',explode=[0.03,0,0])
plt.style.use('ggplot')
plt.show()
```

### As we can see clearly from the piechart that Microsoft is more volatile than the other two companies. So investing in Microsoft would be more risky than others.
