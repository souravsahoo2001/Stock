# Stock analysis of Google

```python
#importing modules
import pandas_datareader.data as pdr
import datetime
import matplotlib.pyplot as plt
```

```python
#Selecting the time to plot a graph using datetime module
start = datetime.datetime(2021,1,1)
end = datetime.datetime(2022,1,1)
```

```python
#Fetching data from yahoo finance website
google = pdr.DataReader("GOOGL", "yahoo", start, end)
```

```python
google.head()
```

## Plot between Open, Close, High and Low price of stock

```python
google['Open'].plot(label = 'Google Open Price',figsize= (15,9))
google['Close'].plot(label = 'Google Close Price')
google['High'].plot(label = 'Google High Price')
google['Low'].plot(label = 'Google Low Price')
plt.legend(loc= 'upper left' )
plt.title('Google Stock Prices', fontsize= 15)
plt.ylabel('Stock Price', fontsize= 15)
plt.xlabel('Date', fontsize= 15)
plt.grid(color = 'black',linestyle= ':' ,linewidth= 0.5)
plt.show()
```

So we can see here in the plot that from the end of october 2021 to november 2021 there is an overall increase in the stock price of google and after that the value went down upto last of november.So here the plot shows that the stock value of google is irregular through out the year.   

```python
google['Volume'].plot(figsize= (15,7))
plt.title('Volume Traded by Google')
```

This graph shows the total numbers of trade occured through out the year.We can clearly see that in the month of February the volume is higher i.e in that month maximum stocks got traded.
